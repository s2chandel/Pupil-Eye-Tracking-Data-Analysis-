```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import butter
from scipy.signal import filtfilt
%matplotlib inline
plt.rcParams["figure.figsize"] = [16,6]
```

```python
pupil1 = pd.read_csv('pupil_positions1.csv')
pupil2 = pd.read_csv('pupil_positions2.csv')
pupil3 = pd.read_csv('pupil_positions3.csv')
```

```python
pupil1['time']= pupil1['timestamp']-pupil1['timestamp'][0] 
pupil2['time']= pupil2['timestamp']-pupil2['timestamp'][0]
pupil3['time']= pupil3['timestamp']-pupil3['timestamp'][0]
```

```python
plt.ylim(0, 1.0)

plt.plot(pupil1['time'],pupil1['norm_pos_x'], label = 'normal1')
plt.plot(pupil2['time'],pupil2['norm_pos_x'], label = 'normal2')
plt.plot(pupil3['time'],pupil3['norm_pos_x'], label = 'normal3')
plt.legend()
```

```python
X_1 = pupil1['norm_pos_x'].mean()
X_2 = pupil2['norm_pos_x'].mean()
X_3 = pupil3['norm_pos_x'].mean()
```

```python
Y_1 = pupil1['norm_pos_y'].mean()
Y_2 = pupil2['norm_pos_y'].mean()
Y_3 = pupil3['norm_pos_y'].mean()
```

```python
plt.ylim(0, 1.0)
plt.xlim(0, 1.0)
plt.plot(X_1,Y_1,marker = 'o',markersize = 15)
plt.plot(X_2,Y_2,marker = 'o',markersize = 15)
plt.plot(X_3,Y_3,marker = 'o',markersize = 15)
```
